## ðŸ”¹ Step 5: Class Distribution

```python
# Value counts for fraud vs non-fraud
class_counts = df['Class'].value_counts()
print(class_counts)

# Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Non-Fraud (0) vs Fraud (1) Transaction Count')
plt.show()

# Class ratio
fraud_ratio = class_counts[1] / class_counts.sum()
print(f"Fraud Ratio: {fraud_ratio:.4%}")
```

#### This means:

#### 284,315 transactions are labeled 0 â†’ These are non-fraudulent (legitimate) transactions.

#### 492 transactions are labeled 1 â†’ These are fraudulent transactions.
